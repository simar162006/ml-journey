```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class label','Alchohol','Malic acid']
```

```python
df
```

```python
sns.kdeplot(df['Alchohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict={1:'red',2:'green',3:'blue'}
sns.scatterplot(x=df['Alchohol'],y=df['Malic acid'],hue=df['Class label'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=0.3,random_state=0)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled= scaler.transform(x_test)
```

```python
x_train_scaled= pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(x_train['Alchohol'],x_train['Malic acid'],c=y_train)
ax1.set_title("Before scaling")
ax2.scatter(x_train_scaled['Alchohol'],x_train['Malic acid'],c=y_train)
ax2.set_title("After scaling")
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
sns.kdeplot(x_train['Alchohol'],ax=ax1)
sns.kdeplot(x_train['Malic acid'],ax=ax1)
ax1.set_title("Before scaling")
sns.kdeplot(x_train_scaled['Alchohol'],ax=ax2)
sns.kdeplot(x_train_scaled['Malic acid'],ax=ax2)
ax2.set_title("After scaling")
plt.show()
```
